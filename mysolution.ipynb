```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from datetime import datetime, timedelta

fifa21_data = pd.read_csv("C:\\Users\\canbe\\OneDrive\\Masaüstü\\codes\\fifa\\fifa21 raw data v2.csv")
fifa21_data.head()
```

```python
fifa21_data['Club'] = fifa21_data['Club'].str.strip() ##Get rid of whitespace in Club column
fifa21_data.rename(columns = {'↓OVA': 'Overall','IR': 'InternationalReputation'}, inplace = True) ##Rename some columns
fifa21_data.drop(columns=['photoUrl', 'playerUrl'], inplace=True) ##Drop unnecessary columns
fifa21_data.head()
```

```python
fifa21_data['Hits'] = pd.to_numeric(fifa21_data['Hits'], errors='coerce') ##Convert Hits column to numeric
fifa21_data.isnull().sum()
```

```python
fifa21_data['Hits'] = fifa21_data['Hits'].fillna(fifa21_data['Hits'].median())
fifa21_data.isnull().sum()
```

```python
fifa21_data['Height'].unique()
fifa21_data['Weight'].unique()
```

```python
def feet_inches_to_cm(height_str):
    if '"' in height_str:
        height_ft_in = height_str.split('\'')
        feet = int(height_ft_in[0])
        inches = int(height_ft_in[1].replace('"', ''))
        return (feet * 12 + inches) * 2.54
    else:
        return float(height_str.replace('cm', ''))

# Convert 'Height' to centimeters
fifa21_data['Height'] = fifa21_data['Height'].apply(feet_inches_to_cm)
fifa21_data['Height'] = fifa21_data['Height'].round().astype(int)
fifa21_data['Height'].unique()
```

```python
def convert_weight(weight_str):
    if 'lbs' in weight_str:
        return float(weight_str.replace('lbs', '')) * 0.453592
    elif 'kg' in weight_str:
        return float(weight_str.replace('kg', ''))
    else:
        return None

# Apply the conversion to kilograms
fifa21_data['Weight'] = fifa21_data['Weight'].apply(convert_weight)
fifa21_data['Weight'] = fifa21_data['Weight'].round().astype(int)
fifa21_data['Weight'].unique()
```

```python
# Convert 'Joined' column to datetime
fifa21_data['Joined'] = pd.to_datetime(fifa21_data['Joined'], format='%b %d, %Y', errors='coerce')

# Extract year, month, and day into separate columns
fifa21_data['Year_Joined'] = fifa21_data['Joined'].dt.year
fifa21_data['Month_Joined'] = fifa21_data['Joined'].dt.month
fifa21_data['Day_Joined'] = fifa21_data['Joined'].dt.day

# Display the new columns to verify the transformation
fifa21_data[['Joined', 'Year_Joined', 'Month_Joined', 'Day_Joined']].head()
```

```python
# Function to convert the 'Value', 'Wage', and 'Release Clause' columns to integer values
def currency_str_to_int(currency_str):
    if pd.isnull(currency_str):
        return None
    if 'M' in currency_str:
        return int(float(currency_str.replace('€', '').replace('M', '')) * 1e6)
    elif 'K' in currency_str:
        return int(float(currency_str.replace('€', '').replace('K', '')) * 1e3)
    else:
        return int(currency_str.replace('€', ''))

# Apply the conversion to the 'Value', 'Wage', and 'Release Clause' columns
fifa21_data['Value'] = fifa21_data['Value'].apply(currency_str_to_int)
fifa21_data['Wage'] = fifa21_data['Wage'].apply(currency_str_to_int)
fifa21_data['Release Clause'] = fifa21_data['Release Clause'].apply(currency_str_to_int)
fifa21_data[['Value', 'Wage', 'Release Clause']].head()
```

```python
fifa21_data['Contract'].head()
```

```python
import re

# Function to extract year from contract string
def extract_year(contract_str):
    if pd.isnull(contract_str):
        return None
    # Find all four consecutive digits which likely represent the year
    years = re.findall(r'\b\d{4}\b', contract_str)
    # If two years are found, return them as start and end, otherwise return None
    return years if len(years) == 2 else [None, None]

# Apply the function to each row in the 'Contract' column
fifa21_data['Contract_Years'] = fifa21_data['Contract'].apply(extract_year)

# Split the list of years into 'Contract Start' and 'Contract End' if available
fifa21_data['Contract Start'] = fifa21_data['Contract_Years'].apply(lambda x: x[0] if x else None)
fifa21_data['Contract End'] = fifa21_data['Contract_Years'].apply(lambda x: x[1] if x else None)

# Drop the intermediate 'Contract_Years' column as it's no longer needed
fifa21_data.drop('Contract_Years', axis=1, inplace=True)

# Display the updated dataframe to verify the results
fifa21_data[['Contract', 'Contract Start', 'Contract End']].head()
```
